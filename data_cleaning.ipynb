# data overview



```python

# ============================================
# KPI CARDS
# ==========================================
```

```python
total_sales = sales_food["sales"].sum()
best_month = sales_food.groupby("month")["sales"].sum().idxmax()
top_industry = sales_food.groupby("industry")["sales"].sum().idxmax()
avg_sales = sales_food["sales"].mean()

print("=" * 40)
print(f"💰 Total Sales:     ${total_sales:,.2f}")
print(f"📅 Best Month:      {best_month}")
print(f"🏆 Top Industry:    {top_industry}")
print(f"📊 Average Sales:   ${avg_sales:,.2f}")
print("=" * 40)
```

```python
# ============================================
# SALES SHARE BY INDUSTRY (% of Total)
# ============================================
```

```python
industry_share = sales_food.groupby("industry")["sales"].sum()
industry_share = (industry_share / industry_share.sum() * 100).round(2).sort_values(ascending=False)
industry_share.name = "sales_share_%"
print(industry_share)
```

```python
top10 = sales_food.groupby("kind_of_business")["sales"].sum().nlargest(10).reset_index()
print(top10)
```

```python
# ============================================
# SALES BY MONTH
# ============================================
month_order = ["January","February","March","April","May","June",
               "July","August","September","October","November","December"]
monthly_sales = (sales_food.groupby("month")["sales"].sum()
                 .reindex([m for m in month_order if m in sales_food["month"].values])
                 .reset_index())
print(monthly_sales)
```

```python
# ============================================
# FOOD vs RETAIL vs AUTOMOTIVE
# ============================================
sales_food["category"] = sales_food["industry"].apply(
    lambda x: "Food" if "Food" in x or "Restaurant" in x
    else "Automotive" if "Automotive" in x
    else "Retail"
)
category_sales = sales_food.groupby("category")["sales"].sum().reset_index()
print(category_sales)
```

```python
# ============================================
# HEATMAP DATA — MONTH vs INDUSTRY
# ============================================
heatmap_data = sales_food.pivot_table(
    values="sales", index="industry", columns="month", aggfunc="sum"
).round(2)
print(heatmap_data)
```

```python

best_months = sales_food.groupby("month")["sales"].sum().sort_values(ascending=False).reset_index()

best_industry_per_month = (
    sales_food.groupby(["month", "industry"])["sales"].sum()
    .reset_index().sort_values(["month", "sales"], ascending=[True, False])
    .groupby("month").first().reset_index()
)

best_business_per_industry = (
    sales_food.groupby(["industry", "kind_of_business"])["sales"].sum()
    .reset_index().sort_values(["industry", "sales"], ascending=[True, False])
    .groupby("industry").first().reset_index()
)

sales_stats = (
    sales_food.groupby(["industry", "kind_of_business", "month"])["sales"]
    .agg(avg_sales="mean", lowest_sales="min", highest_sales="max")
    .reset_index().round(2)
)
```

```python
print("=== BEST MONTHS ===")
print(best_months)

print("\n=== BEST INDUSTRY PER MONTH ===")
print(best_industry_per_month)

print("\n=== BEST BUSINESS PER INDUSTRY ===")
print(best_business_per_industry)

print("\n=== SALES STATS (AVG, MIN, MAX) ===")
print(sales_stats)
```

```python
# ============================================
# SAVE ALL TO CSV
# ============================================
industry_share.reset_index().to_csv("industry_share.csv", index=False)
top10.to_csv("top10_businesses.csv", index=False)
monthly_sales.to_csv("sales_by_month.csv", index=False)
category_sales.to_csv("food_retail_automotive.csv", index=False)
heatmap_data.to_csv("heatmap_month_industry.csv")
best_months.to_csv("best_months.csv", index=False)
best_industry_per_month.to_csv("best_industry_per_month.csv", index=False)
best_business_per_industry.to_csv("best_business_per_industry.csv", index=False)
sales_stats.to_csv("sales_stats.csv", index=False)

print("✅ All CSVs saved!")
```

```python
# ============================================
# KPI CARDS — SAVE TO CSV
# ============================================
kpi_data = {
    "metric": ["Total Sales", "Best Month", "Top Industry", "Average Sales"],
    "value": [f"${total_sales:,.2f}", best_month, top_industry, f"${avg_sales:,.2f}"]
}
kpi_df = pd.DataFrame(kpi_data)
kpi_df.to_csv("kpis.csv", index=False)
print("✅ KPIs saved!")
```

```python
industry_count = {}

for index, row in sales_food.iterrows():
    industry = row["industry"]
    if industry in industry_count:
        industry_count[industry] += 1
    else:
        industry_count[industry] = 1

industry_count_df = pd.DataFrame(
    list(industry_count.items()),
    columns=["industry", "total_records"]
).sort_values("total_records", ascending=False)

print(industry_count_df)
industry_count_df.to_csv("industry_count.csv", index=False)
print("✅ Industry count saved!")
```

```python
industry_count_df = pd.DataFrame(
    list(industry_count.items()),
    columns=["industry", "total_records"]
).sort_values("total_records", ascending=False)

# Add as column to sales_food
sales_food = sales_food.merge(industry_count_df, on="industry", how="left")

print(sales_food.head())
```
